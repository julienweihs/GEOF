## Part 3: Indexing

An **index** describes which part of an array you want to look at or use in estimations. 

Here are a few useful rules to remember about indices in Python:
- Use brackets `[]` around indices. 
- The first index is 0. If you have a list `x=[5, 12, 4, 7, 3, 0]`, then `x[0]=5`.
- To get just the first three elements in `x`, write `x[:3]`. 
- To get just the last two elements in `x`, write `x[4:]` or better `x[-2:]`.
- The last index is not included in the output! If you write `x[2:5]` you choose elements 2 until but NOT including element 5 `x[2:5]=[4, 7, 3]`.

We use the variables from the previous section, `arr1` which is 1D and `arr4` which is 2D as examples below. 

### 1D-indexing

Make a 1D-array

```python
arr1=np.array([5, 12, 4, 7, 3, 0])
```

To find the values from index 3 to the end of the array: 

```python
arr1[3:]
```

To find the first two values:

```python
arr1[:2]
```

If you want the
- first element in `arr1`
- the last element in `arr1`
- the first to and including the second element in `arr1`
- the last three elements in `arr1`

you can write as follows:

```python
print('first element in arr1:', arr1[0])
print('last element in arr1:', arr1[5])
print('first to and including the second element in arr1:', arr1[:2])
print('the three last elements in arr1:', arr1[3:])
```

**Counting backwards**
As briefly mentioned, you can start counting indices at the end of the array and count backwards. This is very practical when working with large arrays and if you want to extract elements at the end of the array. 

The last element has index `-1`, the second to last is `-2`, etc. 

This means that if you want the last element of `arr1` you have two options:

```python
print(arr1)

print(arr1[5]) 
print(arr1[-1])
```

If `arr1` was e.g.,  135423 elemts long it's a lot tidier to write `arr1[-1]` than `arr1[135422]`. With `arr1[-1]` you realize at once that the last element is extracted. With `arr1[135422]` you have to already know that `arr1` has 135423 elemts to realize that the last element is extracted. 

To select the last three elements of `arr1` write:

```python
arr1[-3:]
```

## Part 5: use functions from `Numpy`

To get an overview of the annual temperature in Bergen since 1861, it can be useful to look at values such as average, maximum value, minimum value and standard deviation. `Numpy` has built-in functions for this:

```python
np.mean(temp)
```

```python
print('Average temperature: ', np.mean(temp), '\u2103')
print('Maximum temperature: ', np.max(temp), '\u2103')
print('Minimum temperature: ', np.min(temp), '\u2103')
print('Standard deviation: ', np.std(temp), '\u2103')
```

The new variables `temp` and `time` can be indexed in the same way as in the examples above. In order to calculate the average e.g. of the first 10 years, we do as follows:

```python
np.mean(temp[:10])
```

If we are to calculate the average over the last ten years, we do this as follows:

```python
np.mean(temp[-10:])
```

As a test that this is actually the average of the **ten** first years, you can check the length of `temp[:10]`:

```python
temp[:10].shape
```

## Part 6: Create a figure
Just as we need the `Numpy` library to create matrices and perform mathematical operations such as finding averages, we need a library to create figures. We use `matplotlib`:

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(tid, temp)
```

The figure above reflects what we found by calculating averages: the first 50 years were colder than the last 50 years, the maximum temperature was almost 10 degrees, the minimum temperature was about 6 degrees, and the values mostly stayed below one degree away from the average. But we get much more information additionally, such as *when* it was hottest and coldest, and that the temperature has not increased steadily.

Figures should always be interpretable without a lot of accompanying text - yet we must add titles, names of the axes etc.

```python
plt.plot(tid, temp)
plt.title('Average annual rainfall Bergen')
plt.xlabel('Year')
plt.ylabel('Temperature [\u2103]') # \u2103 is the code for "Celsius degrees"

# The line below delimits the x-axis so that there are no "whitespaces" on
# either ends, as in the figure above
plt.xlim(tid[0], tid[-1])

# The line below just means that information about the previous line is not printed.
# You can comment it out and rerun this cell to see
plt.show()
```

You can create a figure that only shows part of the data set by indexing it in the same way as we have done previously. In order to e.g. only plot the last 100 years, you do it like this:

```python
plt.plot(tid[-100:], temp[-100:])
plt.title('Average annual rainfall Bergen')
plt.xlabel('Year')
plt.ylabel('Temperature [\u2103]')
plt.xlim(tid[-100], tid[-1]) 
plt.show()
```

You can also plot several lines in the same figure. For this, you just add a line with `plt.plot()` for each new variable you want to plot. The plot lines are then created in the order in which they are written.

A small note about lines as we plot the red line for the average value in the cell below: We can write `plt.hlines()` because `plt.hlines()` is a built-in function of `Matplotlib` for plotting horizontal lines. Similarly, there is `plt.vlines()` for vertical lines. We will return to how to find out that such functions exist and how to use them later when we look at documentation. To give a hint: I would just google "plot horizontal line matplotlib" if I didn't remember if there was a function for this or how it worked!

```python
plt.plot(tid, temp) # Plot hele tidsserien
plt.plot(tid[-100:], temp[-100:]) # Plot kun de siste 100 Ã¥rene

# Plot gjennomsnittet som en rÃ¸d linje. 
plt.hlines(np.mean(temp), tid[0], tid[-1], color='red') 
plt.title('Average annual precipitation Bergen')
plt.xlabel('Year')
plt.ylabel('Temperature [\u2103]')
plt.xlim(tid[0], tid[-1]) 
plt.show()
```

The orange line now represents the last 100 years and is plotted on top of the blue line showing the entire time series. The red line shows the average of the entire time series. To create this, we use the function `plt.hlines()` which stands for "horizontal lines". You give the function 1) the y-value the line should cross, 2) the start value of the line, and 3) the end value of the line along the x-axis. In the example above, this is 1) `np.mean(temp)`, 2) `time[0]`, and 3) `time[-1]`.
